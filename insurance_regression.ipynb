Linear regression analysis about how ages (independent variable) affeccts the insurance charges (dependent variable).

```python
#Step 1: Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Step 2: Read data
df=pd.read_csv('insurance.csv')
```

```python
#Step: 3 Inpecting 5 top rows:
df.head()
```

The age and charges columns have both numeric variables. 

```python
#Step: 4 Inspecting if there are null variables.
df.isnull().sum()
```

The analysis above shows that both the age and the chrges columns have no enpty values or data. 

```python
#Step5: Check datafrane info:
df.info()
```

Again this information confirms that there are no null values.


```python
# Step 4: Plot a scatter plot with age on the X-axis and charges on the Y-axis
plt.figure(figsize=(10,6))
plt.scatter(df['age'], df['charges'], alpha=0.5)
plt.title('Scatter plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()

# Step 5: Prepare the data for training
X = df['age'].values.reshape(-1,1)
y = df['charges'].values

# Step 6: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 7: Fit a linear regression model to the data
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Step 8: Make predictions on the test data
y_pred = regressor.predict(X_test)

# Step 9: Plot another scatter plot with the best-fit line
plt.figure(figsize=(10,6))
plt.scatter(X_test, y_test, color='blue', alpha=0.5, label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Best-fit Line')
plt.title('Age vs Charges with Best-fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.grid(True)
plt.show()
```
